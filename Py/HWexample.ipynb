```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import HWfunctions as hw
```

```python
t = np.arange(0.01, 4, 0.01)

th1 = hw.thetaHW(t=t, r=0.5)
th2 = hw.thetaHW(t=t, r=1.0)
th3 = hw.thetaHW(t=t, r=1.5)
```

```python
plt.plot(t, th1, 'k-', label = 'r=0.5')
plt.plot(t, th2, 'b-', label = 'r=1.0')
plt.plot(t, th3, 'r-', label = 'r=1.5')

plt.ylabel(r'$\theta(r,t)$', fontsize=16)
plt.xlabel(r'$t$', fontsize=16)
plt.grid()
plt.legend(fontsize=12)
#plt.savefig('theta.png', bbox_inches='tight')
plt.show()
```

```python
df = pd.read_csv('../tableFGto5.csv')
rho = df['rho'].values
Gexact = df[' G'].values
Fexact = df[' F'].values

G5 = hw.Gexp(rho, n=5); G6 = hw.Gexp(rho, n=6); G7 = hw.Gexp(rho, n=7);
F5 = hw.Fexp(rho, n=5); F6 = hw.Fexp(rho, n=6); F7 = hw.Fexp(rho, n=7);
```

```python
fig, axs = plt.subplots(1,2, figsize=(8,4))

axs[0].plot(rho, np.log10(np.abs(G5 - Gexact)), 'k-', label='n=5')
axs[0].plot(rho, np.log10(np.abs(G6 - Gexact)), 'b-', label='n=6')
axs[0].plot(rho, np.log10(np.abs(G7 - Gexact)), 'r-', label='n=7')
axs[0].set_title(r'$Log_{10}$ Error of G($\rho$)')
axs[0].set_xlabel(r'$\rho$', fontsize=16)
axs[0].set(xlim=[0.5, 2], ylim=[-16, -3])
axs[0].grid()
axs[0].legend(fontsize=12)

axs[1].plot(rho, np.log10(np.abs(F5 - Fexact)), 'k-', label='n=5')
axs[1].plot(rho, np.log10(np.abs(F6 - Fexact)), 'b-', label='n=6')
axs[1].plot(rho, np.log10(np.abs(F7 - Fexact)), 'r-', label='n=7')
axs[1].set_title(r'$Log_{10}$ Error of F($\rho$)')
axs[1].set_xlabel(r'$\rho$', fontsize=16)
axs[1].set(xlim=[0.5, 2], ylim=[-16, -3])
axs[1].grid()
axs[1].legend(fontsize=12)

#plt.savefig('error.png', bbox_inches='tight')
```
