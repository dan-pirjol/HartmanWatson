```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import HWfunctions as hw
```

```python
t = np.arange(0.01, 4, 0.01)

th1 = hw.thetaHW(t=t, r=0.5)
th2 = hw.thetaHW(t=t, r=1.0)
th3 = hw.thetaHW(t=t, r=1.5)
```

```python
plt.plot(t, th1, 'k-', label = 'r=0.5')
plt.plot(t, th2, 'b-', label = 'r=1.0')
plt.plot(t, th3, 'r-', label = 'r=1.5')

plt.ylabel(r'$\theta(r,t)$', fontsize=16)
plt.xlabel(r'$t$', fontsize=16)
plt.grid()
plt.legend(fontsize=12)
#plt.savefig('theta.png', bbox_inches='tight')
plt.show()
```

```python
df = pd.read_csv('../tableFGto5.csv')
rho = df['rho'].values
Gexact = df[' G'].values
Fexact = df[' F'].values

G = hw.Gexp(rho)
F = hw.Fexp(rho)
```

```python
plt.plot(rho, np.log(np.abs(G - Gexact)), 'r-', label = r'Log Error of G($\rho$)')
plt.plot(rho, np.log(np.abs(F - Fexact)), 'b-', label = r'Log Error of F($\rho$)')

plt.ylabel('Log Error', fontsize=16)
plt.xlabel(r'$\rho$', fontsize=16)
plt.grid()
plt.legend(fontsize=12)
#plt.savefig('error.png', bbox_inches='tight')
plt.show()
```
